```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5), sharex=True)

sns.barplot(x=np.array(["superstring", "built-in string"]), y=np.array([259.7, 7480]), palette="deep", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Memory Consumption (in MB)")

sns.barplot(x=np.array(["superstring", "built-in string"]), y=np.array([3.087, 14.703]), palette="deep", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Execution time (in s)")

sns.despine(bottom=True)
plt.tight_layout(h_pad=2)
```
